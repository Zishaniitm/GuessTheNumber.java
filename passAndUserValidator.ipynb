# Password Validation Program


```
import re

def calculate_strength(password, username):
    score = 0

    # +20 if length >= 10
    if len(password) >= 10:
        score += 20

    # +10 for uppercase
    if re.search(r'[A-Z]', password):
        score += 10

    # +10 for lowercase
    if re.search(r'[a-z]', password):
        score += 10

    # +10 for digit
    if re.search(r'\d', password):
        score += 10

    # +10 for at least 1 special character
    if re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
        score += 10

    # +10 for no repeating chars (3+ in a row)
    if not re.search(r'(.)\1{2,}', password):
        score += 10

    # +10 if username not in password
    if username.lower() not in password.lower():
        score += 10

    # +20 if at least 2 DIFFERENT special characters
    special_chars = set(re.findall(r'[!@#$%^&*(),.?":{}|<>]', password))
    if len(special_chars) >= 2:
        score += 20

    return score


def validate_password(password, username):
    errors = []

    # --- Basic Rules ---
    if len(password) < 10:
        errors.append("❌ Must be at least 10 characters long.")

    if not re.search(r'[A-Z]', password):
        errors.append("❌ Must contain at least 1 uppercase letter (A-Z).")

    if not re.search(r'[a-z]', password):
        errors.append("❌ Must contain at least 1 lowercase letter (a-z).")

    if not re.search(r'\d', password):
        errors.append("❌ Must contain at least 1 digit (0-9).")

    if not re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
        errors.append("❌ Must contain at least 1 special character.")

    # --- Advanced Rules ---
    if ' ' in password:
        errors.append("❌ Must not contain spaces.")

    if username.lower() in password.lower():
        errors.append(f"❌ Must not contain the username '{username}'.")

    if re.search(r'(.)\1{2,}', password):
        errors.append("❌ Must not have more than 2 repeating characters in a row (e.g., 'aaa' is invalid).")

    special_chars = set(re.findall(r'[!@#$%^&*(),.?":{}|<>]', password))
    if len(special_chars) < 2:
        errors.append("❌ Must contain at least 2 different special characters.")

    # --- Strength Score ---
    strength = calculate_strength(password, username)
    if strength < 80:
        errors.append(f"❌ Strength score is {strength}/100. Must be at least 80.")

    # --- Result ---
    if not errors:
        print(f"✅ Valid Password!")
        print(f"💪 Strength Score: {strength}/100")
    else:
        print("❌ Invalid Password. Issues found:")
        for err in errors:
            print(f"   {err}")
        print(f"📊 Strength Score: {strength}/100")


# --- Main ---
username = input("Enter username: ")
password = input("Enter password: ")
validate_password(password, username)
```

```
import re


# ---------------- PASSWORD STRENGTH ----------------
def calculate_strength(password, username):
    score = 0

    if len(password) >= 10:
        score += 20

    if re.search(r'[A-Z]', password):
        score += 10

    if re.search(r'[a-z]', password):
        score += 10

    if re.search(r'\d', password):
        score += 10

    if re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
        score += 10

    if not re.search(r'(.)\1{2,}', password):
        score += 10

    if username.lower() not in password.lower():
        score += 10

    special_chars = set(re.findall(r'[!@#$%^&*(),.?":{}|<>]', password))
    if len(special_chars) >= 2:
        score += 20

    return score


# ----------------VALIDATION FUNCTION ----------------
def validate_password(password, username):
    errors = []

    # --- Username Validation ---
    if ' ' in username:
        print("❌ Invalid Username: Must not contain spaces.")
        return

    if not re.match(r'^[A-Za-z0-9_]+$', username):
        print("❌ Invalid Username: Only letters, digits, and underscore (_) allowed.")
        return

    # --- Password Rules ---
    if len(password) < 10:
        errors.append("❌ Must be at least 10 characters long.")

    if not re.search(r'[A-Z]', password):
        errors.append("❌ Must contain at least 1 uppercase letter.")

    if not re.search(r'[a-z]', password):
        errors.append("❌ Must contain at least 1 lowercase letter.")

    if not re.search(r'\d', password):
        errors.append("❌ Must contain at least 1 digit.")

    if not re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
        errors.append("❌ Must contain at least 1 special character.")

    if ' ' in password:
        errors.append("❌ Must not contain spaces.")

    if username.lower() in password.lower():
        errors.append(f"❌ Must not contain the username '{username}'.")

    if re.search(r'(.)\1{2,}', password):
        errors.append("❌ Must not have more than 2 repeating characters in a row.")

    special_chars = set(re.findall(r'[!@#$%^&*(),.?":{}|<>]', password))
    if len(special_chars) < 2:
        errors.append("❌ Must contain at least 2 different special characters.")

    # --- Strength Check ---
    strength = calculate_strength(password, username)

    if strength < 80:
        errors.append(f"❌ Strength score is {strength}/100. Must be at least 80.")

    # --- Result ---
    if not errors:
        print("\n✅ Valid Password!")
        print(f"💪 Strength Score: {strength}/100")
    else:
        print("\n❌ Invalid Password. Issues found:")
        for err in errors:
            print(f"   {err}")
        print(f"\n📊 Strength Score: {strength}/100")


# ---------------- MAIN PROGRAM ----------------
def main():
    username = input("Enter username: ")
    password = input("Enter password: ")
    validate_password(password, username)


if __name__ == "__main__":
    main()
```
